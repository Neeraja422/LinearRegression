```python
import pandas as pd
project_1_dataset = pd.read_csv("../input/project_1_dataset.csv")
project_1_dataset.head()
```

```python
X = project_1_dataset[['YearsExperience']]
y=project_1_dataset[['Salary']]
print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)
```

```python
yhat=lr.predict(X_test)
yhat[0:5]
```

```python
y_test[0:5]
```

```python
print('Coeffienct:' ,lr.coef_)
print('Intercept:',lr.intercept_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train,color='b')
plt.plot(X_train, lr.coef_[0][0]*X_train+ lr.intercept_[0],color='r')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
```

```python
new_predict=lr.predict([[20]])
new_predict
```

```python
from sklearn.metrics import r2_score
r2_score(yhat,y_test)
```
